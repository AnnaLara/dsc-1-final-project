```python
coef = [('bedrooms', -18195.009642388137),
 ('sqft_living', 161.11322332392334),
 ('sqft_basement', -52.705293975683595),
 ('sqft_living15', 72.37624199160486),
 ('WF_1.0', 279274.54076926195),
 ('FL_NUM_1', 22405.635145033553),
 ('FL_NUM_2', 5613.548648406291),
 ('FL_NUM_3', -28019.183793439784),
 ('mi_from_downtown_log', 210431.83668516215),
 ('bath_0-20%', 98775.44016590095),
 ('bath_21-40%', 111239.01947057186),
 ('bath_41-60%', 121021.85143872297),
 ('bath_81-100%', -235988.2372092404),
 ('mi_from_bellevue_log', -109855.7636174455),
 ('mi_from_south_lake_log', -294919.7693134867),
 ('y_intercept', 415562.4254950042)]
```

```python
coef_log = [210431.83668516215, -109855.7636174455, -294919.7693134867]
```

```python
miles_range = np.arange(1, 101)
```

```python
miles_range
```

```python
miles_range_log = np.log(miles_range)
```

```python
miles_range_log
```

```python
# changes in price after changes in mi_from_downtown_log
y_log = 210431.83668516215 * miles_range_log
```

```python
y_log
```

```python
sns.lineplot(miles_range_log, y_log)
```

```python
plt.plot(miles_range_log, y_log)
```

```python
np.log(1)
```

```python
fig, axs = plt.subplots(2, 1, constrained_layout=True)

axs[0].plot(miles_range_log, y_log)
axs[0].set_title('Log transformed variable')
axs[0].set_xlabel('log(mi_from_downtown)')
axs[0].set_ylabel('price')

axs[1].plot(miles_range, y_log)
axs[1].set_title('Raw data')
axs[1].set_xlabel('mi_from_downtown')
axs[1].set_ylabel('price');
plt.savefig('dis_downtown_raw_vs_log.png')
```

```python
y_log_bellevue = -109855.7636174455 * miles_range_log
```

```python
fig, axs = plt.subplots(2, 1, constrained_layout=True)

axs[0].plot(miles_range_log, y_log_bellevue)
axs[0].set_title('Log transformed variable')
axs[0].set_xlabel('log(mi_from_bellevue)')
axs[0].set_ylabel('price')

axs[1].plot(miles_range, y_log_bellevue)
axs[1].set_title('Raw data')
axs[1].set_xlabel('mi_from_bellevue')
axs[1].set_ylabel('price');
plt.savefig('dis_bellvue_raw_vs_log.png')
```

```python
y_log_s_lake_union = -294919.7693134867 * miles_range_log
```

```python
fig, axs = plt.subplots(2, 1, constrained_layout=True)

axs[0].plot(miles_range_log, y_log_s_lake_union)
axs[0].set_title('Log transformed variable')
axs[0].set_xlabel('log(mi_from_s_lake_union)')
axs[0].set_ylabel('price')

axs[1].plot(miles_range, y_log_s_lake_union)
axs[1].set_title('Raw data')
axs[1].set_xlabel('mi_from_s_lake_union')
axs[1].set_ylabel('price');
plt.savefig('dis_south_late_union_raw_vs_log.png')
```
